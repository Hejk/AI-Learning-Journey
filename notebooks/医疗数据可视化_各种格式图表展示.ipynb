```python
import pandas as pd
data = pd.read_csv('某地乳腺检查数据.csv')
data.head()
```

```python
y=data['诊断结果']  #标签集 为y轴
X = data.drop(['诊断结果','ID'],axis = 1) #特征集 删除ID诊断结果列,axis =1 代表删除列 ; axis =0 代表删除行
X  #输出 X
```

用柱状图显示数据
代码如下：

```python
import matplotlib.pyplot as plt #引入matplotlib.pyplot工具包
#设置中文字体显示
plt.rcParams["font.family"]=["SimHei"]
plt.rcParams['font.sans-serif']=['SimHei'] 
# 用来正常显示负号
plt.rcParams['axes.unicode_minus']=False

#显示y的柱状图
y.value_counts().plot(kind='bar',color='skyblue', edgecolor='black') 
plt.title('诊断结果分布')
plt.xlabel('诊断结果')
plt.ylabel('确诊数量')
plt.xticks(rotation=45) #x轴的刻度倾斜45°
plt.tight_layout() #自动调整子图/图表的边距和布局，使所有元素（标题、标签、刻度等）完整显示在画布内
plt.show()
```

```python
X.describe() #显示X轴信息
```

```python
#查看X轴的特征名称
features = X.columns
features
```

绘制前三个特征的直方图

```python
first_three_features = features[:3] #提取前3个特征

#创建画布对象
#fig代表整个画布,axes包含所有子图轴的数组
#nrows=3,ncols=1 生成 3 行 1 列（共 3 个子图）的布局 
#figsize=(8, 10)：画布尺寸（宽 8 英寸，高 10 英寸）
fig,axes = plt.subplots(nrows=3,ncols=1,figsize=(8,10)) 

#enumerate() 是 Python 中的一个内置函数，用于 给可迭代对象（如列表、元组、字符串等）添加序号，使你在遍历时能同时获取元素的 索引（位置） 和 值
for i,feature in enumerate(first_three_features):
    axes[i].hist(X[feature],bins=30,color='blue',edgecolor='black')
    axes[i].set_title(f"{feature} - 直方图")
    axes[i].set_xlabel('值')
    axes[i].set_ylabel('频数')
    axes[i].grid(True,axis='y',linestyle='--',linewidth=0.7,alpha=0.6)

plt.tight_layout()
plt.show()
```

绘制箱线图

```python
#选择前10个特征绘制箱线图
selected_features = features[:10]
selected_features
```

```python
plt.figure(figsize=(12,8)) # 创建画布，尺寸为12x8英寸
plt.boxplot([X[feature] for feature in selected_features],vert=True) # 画垂直箱线图
plt.xticks(range(1,len(selected_features)+1),labels=selected_features,rotation=45) #第三个参数应为关键字参数,如果直接传参会报错~
plt.title("箱线图-前10个特征") 
plt.ylabel('值')
plt.grid(True,axis='y',linestyle='--',linewidth=0.7,alpha=0.6)
plt.tight_layout() #避免遮挡：调用 plt.tight_layout() 确保所有元素完整显示
plt.show()
```

引入sklearn.preprocessing 标准化工具 StandardScaler

```python
from sklearn.preprocessing import StandardScaler

#setting x and y
y=data['诊断结果']
X = data.drop(['诊断结果','ID'],axis=1)

selected_features = X.columns[:10]

scaler = StandardScaler() # 实例化标准化器
X_standardized =scaler.fit_transform(X) # 拟合数据并转换

X_selected_standardized = X_standardized[:,:10] # 选取前10列（对应selected_features）
```

```python
import matplotlib.pyplot as plt

# Setting the font for displaying Chinese characters in the plot
plt.rcParams["font.family"]=['SimHei']
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(12,8)) # 创建画布，尺寸为12x8英寸
plt.boxplot(X_selected_standardized,vert=True) # 画垂直箱线图  X_selected_standardized为标准化后的数据（每个特征为箱线图的一列）
plt.xticks(range(1,len(selected_features)+1),labels=selected_features,rotation=45) #第三个参数应为关键字参数,如果直接传参会报错~
plt.title("箱线图-标准化后的前10个特征") 
plt.ylabel('标准化值')
plt.grid(True,axis='y',linestyle='--',linewidth=0.7,alpha=0.6)
plt.tight_layout() #避免遮挡：调用 plt.tight_layout() 确保所有元素完整显示
plt.show()
```

绘制小提琴图

```python
import seaborn as sns #导入Seaborn
#绘制小提琴图
plt.figure(figsize=(12,8)) #创建画布

# 转换为DataFrame并指定列名
# 使用"Set3"调色板着色
sns.violinplot(data=pd.DataFrame(X_selected_standardized,columns=selected_features),palette="Set3") 
plt.title("小提琴图-前10个特征") 
plt.ylabel('标准化值')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

绘制热点图（热力图）

```python
#计算相关系数矩阵

correlation_matrix = pd.DataFrame(X_selected_standardized,columns=selected_features).corr()
plt.figure(figsize=(12,8))
sns.heatmap(
    correlation_matrix,   # 输入相关系数矩阵
    annot=True,           # ✅在单元格内显示数值
    cmap='coolwarm',      # ✅使用冷-暖色系（蓝-红渐变）
    linewidths=0.5        # ✅设置单元格间线条宽度为0.5
)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
